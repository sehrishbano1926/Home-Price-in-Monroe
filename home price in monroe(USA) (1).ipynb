# Vedio 2(Hindi Machine Learning Tutorial 2 - Linear Regression Single Variable)


```python
home_price={
    'Area':['2600', '3000', '3200', '3600', '4000'],
    'Price':['550000', '565000', '610000', '680000', '725000']
}

home_price_df=pd.DataFrame(home_price)
home_price_df
    
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(home_price_df['Area'],home_price_df['Price'],color='red',marker='+')
plt.show()
```

# Drop Price column

when the model of object is ready and we call the method, linear_model expect 2d array.
that is why we drop price columne

```python
new_home_price_df=home_price_df.drop('Price', axis='columns')
new_home_price_df
```

# Predict the price of  home when the area is given????

first we will find the price of home having area of 3300..
after that, apply the same method to find home price having area of 5000

```python
model=LinearRegression()
model.fit(new_home_price_df, home_price['Price'] )
```

```python
model.predict([[3300]])
```

```python
model.predict([[5000]])
```
